```python
import pandas as pd
from matplotlib import pyplot
from fbprophet import Prophet


car = pd.read_csv('서울전기차.csv')
car.columns = ['ds','y']
car['ds'] = pd.to_datetime(car['ds'])

model = Prophet()

model.fit(car)

last_1year = list()
for i in range(3,13):
    last_1year.append(['2021-%02d' % i])
last_1year = pd.DataFrame(last_1year,columns = ['ds'])
last_1year['ds'] = pd.to_datetime(last_1year['ds'])
forecast = model.predict(last_1year)
model.plot(forecast)
pyplot.show()

future_1year = list()
for i in range(2, 13):
    future_1year.append(['2022-%02d' % i])
future_1year = pd.DataFrame(future_1year, columns = ['ds'])
future_1year['ds']= pd.to_datetime(future_1year['ds'])

forecast1 = model.predict(future_1year)
model.plot(forecast1)
pyplot.show()
```
